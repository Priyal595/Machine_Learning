```python
import numpy as np

# Get all symptom columns
symptom_cols = [col for col in df.columns if col.startswith('Symptom')]

# Get unique symptoms
all_symptoms = set()

for col in symptom_cols:
    all_symptoms.update(df[col].dropna().unique())

all_symptoms = list(all_symptoms)

# Create empty dataframe
X = pd.DataFrame(0, index=df.index, columns=all_symptoms)

# Fill with 1 where symptom exists
for i, row in df.iterrows():
    for col in symptom_cols:
        symptom = row[col]
        if pd.notna(symptom):
            X.at[i, symptom] = 1

# Target
y = df['Disease']
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
X_unique = pd.DataFrame(X).drop_duplicates()
y_unique = y[:len(X_unique)]
```

```python
X_unique.shape
```

```python
df_unique = pd.concat([pd.DataFrame(X), pd.Series(y, name="target")], axis=1)
df_unique = df_unique.drop_duplicates()

X_unique = df_unique.drop("target", axis=1).values
y_unique = df_unique["target"].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_unique, y_unique, test_size=0.2, random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import numpy as np
from collections import Counter

class KNN:
    def __init__(self, k=5):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def distance(self, x1, x2):
        return np.sqrt(np.sum((x1 - x2) ** 2))

    def predict(self, X):
        predictions = []

        for x in X:
            distances = [self.distance(x, x_train) for x_train in self.X_train]

            k_indices = np.argsort(distances)[:self.k]
            k_labels = [self.y_train[i] for i in k_indices]

            most_common = Counter(k_labels).most_common(1)
            predictions.append(most_common[0][0])

        return np.array(predictions)
```

```python
knn = KNN(k=5)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
```

```python
def accuracy(y_true, y_pred):
    return np.mean(y_true == y_pred)

print("Accuracy:", accuracy(y_test, y_pred))
```

```python
import numpy as np

class ImprovedKNN:
    def __init__(self, base_k=5):
        self.base_k = base_k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

        # Feature weights (equal for now, can be customized later)
        self.weights = np.ones(X.shape[1])
        self.weights = self.weights / np.sum(self.weights)

    # 🔹 Weighted Euclidean Distance
    def weighted_distance(self, x1, x2):
        return np.sqrt(np.sum(self.weights * (x1 - x2) ** 2))

    # 🔹 Adaptive K + Weighted Voting
    def predict(self, X):
        predictions = []
        confidences = []

        for x in X:
            # Compute distances
            distances = np.array([
                self.weighted_distance(x, x_train)
                for x_train in self.X_train
            ])

            # 🔥 Adaptive K (based on spread of distances)
            spread = np.std(distances)
            k = int(self.base_k + spread * 2)
            k = max(3, min(k, 15))  # keep k in range

            # Get K nearest neighbors
            k_indices = np.argsort(distances)[:k]
            k_distances = distances[k_indices]
            k_labels = self.y_train[k_indices]

            # 🔥 Distance-weighted voting
            votes = {}
            for label, dist in zip(k_labels, k_distances):
                weight = 1 / (dist + 1e-5)  # avoid division by zero
                votes[label] = votes.get(label, 0) + weight

            # Prediction
            pred = max(votes, key=votes.get)

            # 🔥 Confidence score
            confidence = votes[pred] / sum(votes.values())

            predictions.append(pred)
            confidences.append(confidence)

        return np.array(predictions), np.array(confidences)
```

```python
model = ImprovedKNN(base_k=5)
model.fit(X_train, y_train)

y_pred_improved, conf = model.predict(X_test)

print("Improved Accuracy:", accuracy(y_test, y_pred_improved))
print("Average Confidence:", np.mean(conf))
```
